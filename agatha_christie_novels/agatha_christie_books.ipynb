```python
# import libraries 

from bs4 import BeautifulSoup
import requests
import time
import datetime
import re
```

```python
#get webpage content
url='https://en.wikipedia.org/wiki/Agatha_Christie_bibliography#Novels'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", 
                  "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
response=requests.get(url,headers)
time.sleep(1)

soup1 = BeautifulSoup(response.text, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
```

```python
table=soup2.select('.wikitable.sortable')[0]


col_list=table.find('tr').find_all('th')[1:]

column=[col.get_text(strip=True)for col in col_list]

rows=table.find_all('tr')[1:]
table_info_list=[]
for r in rows:
    table_info={}
    table_info['title']=r.find('i').a.get_text(strip=True)
    table_info['link']='https://en.wikipedia.org/'+r.a['href']
    content_value=[c.get_text(strip=True) for c in r.find_all('td')]
    
    result= dict(zip(column,content_value))
    table_info.update(result)
    table_info_list.append(table_info)

table_info_list
                      
        
```

```python
import json
def save_data(title, data):
    with open(title, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

def load_data(title):
    with open(title, encoding="utf-8") as f:
        return json.load(f)
```

```python
save_data('agatha_christie_novels',table_info_list)
```

```python
load_data('agatha_christie_novels')
```
